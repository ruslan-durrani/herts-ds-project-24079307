# EDA

First, we import the libraries we need.
These libraries help us read the data, clean it, and create visualisations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile
import os
import glob

pd.set_option("display.max_columns", None)
plt.rcParams["figure.figsize"] = (10, 6)
sns.set_style("whitegrid")
```

Extract the Uploaded ZIP File

```python
zip_path = "./data.zip"
extract_path = "extracted_data"

os.makedirs(extract_path, exist_ok=True)

with zipfile.ZipFile(zip_path, "r") as z:
    z.extractall(extract_path)

print("Extracted to:", extract_path)
```

Find all csv files

```python
all_csv_files = glob.glob(os.path.join(extract_path, "**", "*.csv"), recursive=True)
print("CSV files found:", len(all_csv_files))
all_csv_files[:5]
```

Load all CSV files and combine them

```python
dfs = []
for f in all_csv_files:
    temp = pd.read_csv(f)
    dfs.append(temp)

df = pd.concat(dfs, ignore_index=True)
print("Combined shape:", df.shape)
df.head()
```

Check required columns exist

```python
required_cols = ["Month", "Falls within", "Latitude", "Longitude", "LSOA code", "Crime type", "Last outcome category"]
missing_cols = [c for c in required_cols if c not in df.columns]
print("Missing required columns:", missing_cols)
```

Confirm all 3 police forces are present.

```python
forces = sorted(df["Falls within"].dropna().unique().tolist())
forces
```

Quick count per police force (to see balance)

```python
df["Falls within"].value_counts()
```

Check data types and missing values

```python
df.info()

missing_pct = (df.isnull().mean() * 100).round(2).sort_values(ascending=False)
missing_pct
```

Convert Month to datetime and create time features.

```python
df["Month"] = pd.to_datetime(df["Month"], errors="coerce")
df = df.dropna(subset=["Month"])  # drop only if Month is invalid

df["year"] = df["Month"].dt.year
df["month_num"] = df["Month"].dt.month

df[["Month", "year", "month_num"]].head()
```

**Street-level to LSOA-month aggregation**


Aggregate to LSOA-month crime counts.

```python
agg_lsoa = (
    df.groupby(["LSOA code", "year", "month_num"])
      .size()
      .reset_index(name="crime_count")
)

agg_lsoa.head()
```

Add average latitude and longitude per LSOA.

```python
lsoa_spatial = (
    df.groupby("LSOA code")
      .agg({"Latitude": "mean", "Longitude": "mean"})
      .reset_index()
)

agg_lsoa = agg_lsoa.merge(lsoa_spatial, on="LSOA code", how="left")
agg_lsoa.head()
```

Save the aggregated dataset.

```python
os.makedirs("outputs", exist_ok=True)
agg_lsoa.to_csv("outputs/aggregated_lsoa_month.csv", index=False)
print("Saved:", "outputs/aggregated_lsoa_month.csv", "rows:", len(agg_lsoa))
```

```python
agg_lsoa["log_crime"] = np.log1p(agg_lsoa["crime_count"])
```

# Hotspot Detection Using Clustering

Prepare spatial and temporal features for clustering.

```python
from sklearn.preprocessing import StandardScaler

X_cluster = agg_lsoa[["log_crime", "Latitude", "Longitude", "month_num"]]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)
```

Apply KMeans clustering.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=4, random_state=42)
k_labels = kmeans.fit_predict(X_scaled)

agg_lsoa["kmeans_cluster"] = k_labels

sil_k = silhouette_score(X_scaled, k_labels)
print("KMeans Silhouette Score:", sil_k)
```

Apply DBSCAN clustering.

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.8, min_samples=10)
d_labels = dbscan.fit_predict(X_scaled)

agg_lsoa["dbscan_cluster"] = d_labels

if len(set(d_labels)) > 1:
    sil_d = silhouette_score(X_scaled, d_labels)
    print("DBSCAN Silhouette Score:", sil_d)
```

Visualise KMeans spatial clusters.

```python
sns.scatterplot(
    x="Longitude",
    y="Latitude",
    hue="kmeans_cluster",
    data=agg_lsoa,
    palette="tab10",
    s=15
)

plt.title("Spatial Hotspots - KMeans")
plt.show()
```

check how many clusters were formed and how many points labeled -1 (noise)

```python
pd.Series(agg_lsoa["dbscan_cluster"]).value_counts()
```

Are some clusters actually higher crime?

```python
agg_lsoa.groupby("dbscan_cluster")["crime_count"].mean()
```

**So DBSCAN demonstrates stronger capability in identifying extreme high-risk hotspots compared to KMeans, as evidenced by a distinct cluster with substantially higher mean crime intensity.**

Are they mostly from one police force?

```python
agg_lsoa[agg_lsoa["dbscan_cluster"] == 2]["LSOA code"].nunique()
```

```python
agg_lsoa[agg_lsoa["dbscan_cluster"] == 2].head()
```

```python
df.merge(agg_lsoa[["LSOA code", "dbscan_cluster"]], on="LSOA code") \
  .query("dbscan_cluster == 2")["Falls within"].value_counts()
```

**DBSCAN is more effective at identifying extreme high-risk crime hotspots due to its density-based clustering mechanism.**

**KMeans grouped areas more broadly but did not isolate extreme hotspots as clearly.**

# Crime Forecasting

Create crime type frequency feature per LSOA-month

```python
# count total crimes per LSOA-month (already have crime_count)
# now compute dominant crime type frequency ratio

crime_type_counts = (
    df.groupby(["LSOA code", "year", "month_num", "Crime type"])
      .size()
      .reset_index(name="type_count")
)

# find dominant crime type per LSOA-month
dominant_type = (
    crime_type_counts.sort_values("type_count", ascending=False)
      .drop_duplicates(["LSOA code", "year", "month_num"])
)

dominant_type = dominant_type.rename(columns={"type_count": "dominant_type_count"})

dominant_type = dominant_type[["LSOA code", "year", "month_num", "dominant_type_count"]]
```

Merge crime composition feature into LSOA dataset.

```python
forecast_df = agg_lsoa.merge(
    dominant_type,
    on=["LSOA code", "year", "month_num"],
    how="left"
)
```

Create seasonal indicator and lag feature.

```python
# seasonal indicator (winter=1, else=0)
forecast_df["is_winter"] = forecast_df["month_num"].isin([12, 1, 2]).astype(int)

# sort properly
forecast_df = forecast_df.sort_values(["LSOA code", "year", "month_num"])

# lag feature
forecast_df["lag_1"] = forecast_df.groupby("LSOA code")["crime_count"].shift(1)

forecast_df = forecast_df.dropna()
```

Encode LSOA as numeric feature.

```python
forecast_df["lsoa_encoded"] = forecast_df["LSOA code"].astype("category").cat.codes
```

Define full feature set and target variable.

```python
X_features = [
    "lag_1",
    "year",
    "month_num",
    "is_winter",
    "Latitude",
    "Longitude",
    "dominant_type_count",
    "lsoa_encoded"
]

X = forecast_df[X_features]
y = forecast_df["crime_count"]

print("Feature shape:", X.shape)
```

Split data using time-based split.

```python
train = forecast_df[forecast_df["year"] < 2025]
test = forecast_df[forecast_df["year"] >= 2025]

X_train = train[X_features]
y_train = train["crime_count"]

X_test = test[X_features]
y_test = test["crime_count"]
```

Train Linear Regression model




```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

lr = LinearRegression()
lr.fit(X_train, y_train)

pred_lr = lr.predict(X_test)

rmse_lr = np.sqrt(mean_squared_error(y_test, pred_lr))
mae_lr = mean_absolute_error(y_test, pred_lr)

print("Linear Regression RMSE:", rmse_lr)
print("Linear Regression MAE:", mae_lr)
```

Train Random Forest model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

pred_rf = rf.predict(X_test)

rmse_rf = np.sqrt(mean_squared_error(y_test, pred_rf))
mae_rf = mean_absolute_error(y_test, pred_rf)

print("Random Forest RMSE:", rmse_rf)
print("Random Forest MAE:", mae_rf)
```

Compare both models

```python
results = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest"],
    "RMSE": [rmse_lr, rmse_rf],
    "MAE": [mae_lr, mae_rf]
})

results
```

Overfitting Check

```python
train_pred = rf.predict(X_train)

train_rmse = np.sqrt(mean_squared_error(y_train, train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, pred_rf))

print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
```

**The Random Forest model demonstrates overfitting, as evidenced by a substantial gap between training and testing error. The model captures complex spatial and temporal relationships but does not fully generalise to unseen time periods.**

Apply regularisation to Random Forest by constraining tree depth and leaf size to improve generalisation.

```python
rf_reg = RandomForestRegressor(
    n_estimators=200,
    max_depth=12,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)

rf_reg.fit(X_train, y_train)

pred_rf_reg = rf_reg.predict(X_test)

train_rmse_reg = np.sqrt(mean_squared_error(y_train, rf_reg.predict(X_train)))
test_rmse_reg = np.sqrt(mean_squared_error(y_test, pred_rf_reg))

print("Regularised Train RMSE:", train_rmse_reg)
print("Regularised Test RMSE:", test_rmse_reg)
```

**The regularised Random Forest model reduces overfitting by constraining tree depth and leaf size. The gap between training and testing error decreases substantially, indicating improved generalisation while maintaining similar predictive performance on unseen data.**

# Feature Selection

Select the most important features using statistical feature selection.

```python
from sklearn.feature_selection import SelectKBest, f_regression

selectorAll = SelectKBest(score_func=f_regression, k="all")
selectorAll.fit(X_train, y_train)

featureScores = pd.DataFrame({
    "Feature": X_train.columns,
    "Score": selectorAll.scores_
}).sort_values("Score", ascending=False)

featureScores
```

Select top 5 features based on highest scores.

```python
selectorTop5 = SelectKBest(score_func=f_regression, k=5)

XTrainTop5 = selectorTop5.fit_transform(X_train, y_train)
XTestTop5 = selectorTop5.transform(X_test)

top5Features = X_train.columns[selectorTop5.get_support()]

print("Top 5 Features:", list(top5Features))
```

Train regularised Random Forest using selected top features.

```python
rfTop5 = RandomForestRegressor(
    n_estimators=200,
    max_depth=12,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)

rfTop5.fit(X_train[top5Features], y_train)

predTop5 = rfTop5.predict(X_test[top5Features])

rmseTop5 = np.sqrt(mean_squared_error(y_test, predTop5))

print("RMSE using Top 5 Features:", rmseTop5)
```

**Feature selection results indicate that temporal lag and crime composition are the most influential predictors of future crime intensity, while long-term year trends contribute minimally.**

# SHAP Interpretability

Use SHAP to explain how each feature contributes to crime prediction.

```python
import shap

# create explainer
explainer = shap.TreeExplainer(rfTop5)

# calculate SHAP values
shapValues = explainer.shap_values(X_test[top5Features])
```

Visualise global feature importance using SHAP summary plot.

```python
shap.summary_plot(shapValues, X_test[top5Features])
```

Visualise average feature importance.

```python
shap.summary_plot(shapValues, X_test[top5Features], plot_type="bar")
```

**SHAP analysis confirms that crime composition and previous month crime intensity are the most influential predictors of future crime levels. Spatial location also contributes to prediction, indicating that crime risk is both temporally persistent and geographically structured.**
